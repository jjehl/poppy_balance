```python
from poppy.creatures import PoppyHumanoid

poppy = PoppyHumanoid(simulator='vrep')

%pylab inline
import time
```

Test de la vitesse de rafraichissement de V-REP. L

enregistrement de la chute apr√©s un lever de pied

```python
poppy.l_shoulder_x.goto_behavior = 'minjerk'

poppy.get_object_position('head_visual')

pos_x = []
pos_y = []
pos_z = []
t= []

t0 = time.time()

poppy.r_hip_y.goto_position(-50, 4)
poppy.r_knee_y.goto_position(50, 4)


while time.time() - t0 < 30:
    pos = poppy.get_object_position('head_visual')
    pos_x.append(pos[0])
    pos_y.append(pos[1])
    pos_z.append(pos[2])
    t.append(poppy.current_simulation_time)
    time.sleep(0.02)
    
plot(t, pos_x)
plot(t, pos_y)
plot(t, pos_z)


legend(('x', 'y','z'))
```

```python
poppy.l_shoulder_x.goto_behavior = 'minjerk'

list_pos_x = []
list_pos_y = []
list_pos_z = []
t= []

t0 = time.time()

poppy.r_hip_y.goto_position(-50, 4)
poppy.r_knee_y.goto_position(50, 4)

pos = poppy.get_object_position('head_visual')
pos_x=pos[0]
pos_y=pos[1]
pos_z=pos[2]

while time.time() - t0 < 30:
    pos = poppy.get_object_position('head_visual')
    
    decalage_x=pos_x-pos[0]
    decalage_y=pos_y-pos[1]
    decalage_z=pos_z-pos[2]

    list_pos_x.append(decalage_x)
    list_pos_y.append(decalage_y)
    list_pos_z.append(decalage_z)
    
    pos_x = pos[0]
    pos_y = pos[1]
    pos_z = pos[2]
    
    t.append(poppy.current_simulation_time)
    #poppy.abs_x.goal_position=
    time.sleep(0.05)
    
plot(t, list_pos_x)
plot(t, list_pos_y)
plot(t, list_pos_z)


legend(('dx', 'dy','dz'))
```
